```python
# Importing all the necessary libraries required to Scrape data using Selenium
import selenium
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException,NoSuchAttributeException 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import re
```

```python
driver1 = webdriver.Chrome(r'G:\AI Professional\Data Trained\Internship Fliprobo Technologies\Webscraping (Selenium)\chromedriver.exe')
```

```python
driver1.maximize_window()
```

```python
driver1.get('https://www.amazon.in/')
```

```python
# list to store URL's of all the products
product_url = []
```

```python
items = ['laptops', 'Phones', 'Headphones', 'smart watches', 'Professional Cameras', 
         'Printers', 'Monitors', 'Home theater', 'Router', 'bags', 'hard disk', 
         'handbags','TV','books', 'men shoes', 'men shirt', 'men watches', 'women watches', 
         'men jackets', 'women jacket', 'jewellry', 'tablets', 'keyboard', 'wired mouse', 'wireless mouse']
for i in items:
    # Entering the item to be searched
    product = driver1.find_element(By.XPATH, '//input[@type="text"]')
    product.send_keys(i)
    
    # clicking the search button 
    search = driver1.find_element(By.XPATH, '//div[@class="nav-search-submit nav-sprite"]')
    search.click()
    
    # Getting the url's of each product in each item only on page one
    urls = driver1.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for a in urls:
        product_url.append(a.get_attribute('href'))
        
    # clearing the search field for new value
    driver1.find_element(By.XPATH, '//input[@type="text"]').clear()
```

```python
len(product_url)
```

```python
prod = []
review = []
rating = []


for u in product_url:
    driver1.get(u)
    time.sleep(2)
    
    try:
        all_reviews = driver1.find_element(By.XPATH, '//a[@class="a-link-emphasis a-text-bold"]')
        all_reviews.click()
        time.sleep(1)
    except NoSuchElementException:
        continue

    rev = driver1.find_elements(By.XPATH, '//a[@class="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"]/span')
    for r in rev:
        review.append(r.text)

    rtg = driver1.find_elements(By.XPATH, '//div[@class="a-section celwidget"]/div[2]/a')
    for rt in rtg:
        if rt.get_attribute('title') == '':
            continue
        else:
            rating.append(float(rt.get_attribute('title').split()[0]))

        prod1 = driver1.find_element(By.XPATH, '//h1[@class="a-size-large a-text-ellipsis" or @class="a-size-large"]')
        prod.append(prod1.text.split(sep = '/')[0])


print(prod)
print(review)
print(rating)
```
