##### Problem Statement:
Avocado is a fruit consumed by people heavily in the United States. 

##### Content
This data was downloaded from the Hass Avocado Board website in May of 2018 & compiled into a single CSV. 

The table below represents weekly 2018 retail scan data for National retail volume (units) and price. Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Hass avocados. 

Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags. 

The Product Lookup codes (PLU’s) in the table are only for Hass avocados. Other varieties of avocados (e.g. greenskins) are not included in this table.

##### Some relevant columns in the dataset:

Date - The date of the observation
AveragePrice - the average price of a single avocado
type - conventional or organic
year - the year
Region - the city or region of the observation
Total Volume - Total number of avocados sold
4046 - Total number of avocados with PLU 4046 sold
4225 - Total number of avocados with PLU 4225 sold
4770 - Total number of avocados with PLU 4770 sold

##### Splitting the data

### Model Building

#### Regression Analysis

```python
# Importing the libraries for model building
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiating each algorithm
lr = LinearRegression()
dtr = DecisionTreeRegressor()
knr = KNeighborsRegressor()
svr = SVR()
rfr = RandomForestRegressor()

# List of instantiated algorithms
model = [lr, dtr, knr, svr, rfr]
```

```python
# Maximum accuracy
maxAccu_av1 = 0 

# Best random state value for which accuracy is achieved 
maxRS_av1 = 0

for m in model:
    for i in range(0,70):
        x_train, x_test,y_train, y_test = train_test_split(X1,y1, random_state=i, test_size=0.20)
        m.fit(x_train, y_train)
        pred_train_av1= m.predict(x_train)
        pred_test_av1 = m.predict(x_test)
        score_train_av1 = m.score(x_train, y_train)
        score_test_av1 = m.score(x_test, y_test)
        r2_av1 = r2_score(y_test, pred_test_av1)
        print(f'----------------At random State {i} ------------------')
        print(f"Training score is {score_train_av1}")
        print(f"Testing score is {score_test_av1}")
        print(f'r2 score of {m} is {r2_av1}')
    
        if r2_av1 > maxAccu_av1:
            maxAccu_av1 = r2_av1
            maxRS_av1 = i
            print(f"Maximum score of {m} is {maxAccu_av1} at Random state {i}")
            print('\n')
    print('*'*120)
```

- Getting Maximum accuracy for Random forest of 0.9091012819228353 at Random state 28

##### Cross validation of all the Algorithms for regression

```python
# Before proceeding with the hpyer parameter tunning we will check for cross validation
from sklearn.model_selection import cross_val_score
max_cvscore_av1 = 0 
max_cv_av1 = 0
for m in model:
    print(f'---------------------- Cross Validation of {m} --------------------------')
    for j in range(2,5):
        cv_score_av1 = cross_val_score(m, X1,y1,cv=j)
        cv_mean_av1 = cv_score_av1.mean()
        print(f"At cross fold {j} the cv score is {cv_mean_av1}")
        print('\n') 
    
        if cv_mean_av1 > max_cvscore_av1:
            max_cvscore_av1=cv_mean_av1
            max_cv_av1 = j
            print(f"At cross fold {j} the Maximum CV score is {max_cvscore_av1}")
            print('\n') 
    print('*'*120)
    print('\n')
```

```python
# Training Random forest Regressor at Random state 28
x_train, x_test,y_train, y_test = train_test_split(X1,y1, random_state=28, test_size=0.20)
rfr.fit(x_train, y_train)
pred_train_av1= rfr.predict(x_train)
pred_test_av1 = rfr.predict(x_test)
score_train_av1 = rfr.score(x_train, y_train)
score_test_av1 = rfr.score(x_test, y_test)
r2_av1 = r2_score(y_test, pred_test_av1)
print(f"Training score is {score_train_av1}")
print(f"Testing score is {score_test_av1}")
print(f'r2 score of {rfr} is {r2_av1}')
```

### Hyper Parameter Tunning

```python
from sklearn.model_selection import GridSearchCV
param = {'n_estimators':[100],'max_depth': [2,4],'criterion': ['mse','mae'],'max_features':['sqrt','log2']}

gcv1 = GridSearchCV(rfr,param,cv=4,scoring= 'accuracy')

# training
gcv1.fit(x_train, y_train)

# Getting the best parameters
print("Best parameters:", gcv1.best_params_)

print("Best Estimator:",gcv1.best_estimator_)

gcv_pred_av1=gcv1.best_estimator_.predict(x_test)

print('Final Accuracy with Linear Regression:', r2_score(y_test,gcv_pred_av1))
```
