##### Problem Statement:
Every year a lot of companies hire a number of employees. The companies invest time and money in training those employees, not just this but there are training programs within the companies for their existing employees as well. The aim of these programs is to increase the effectiveness of their employees. But where HR Analytics fit in this? and is it just about improving the performance of employees?

##### HR Analytics

Human resource analytics (HR analytics) is an area in the field of analytics that refers to applying analytic processes to the human resource department of an organization in the hope of improving employee performance and therefore getting a better return on investment. HR analytics does not just deal with gathering data on employee efficiency. Instead, it aims to provide insight into each process by gathering data and then using it to make relevant decisions about how to improve these processes.

##### Attrition in HR

Attrition in human resources refers to the gradual loss of employees overtime. In general, relatively high attrition is problematic for companies. HR professionals often assume a leadership role in designing company compensation programs, work culture, and motivation systems that help the organization retain top employees.

How does Attrition affect companies? and how does HR Analytics help in analyzing attrition? We will discuss the first question here and for the second question, we will write the code and try to understand the process step by step.

##### Attrition affecting Companies

A major problem in high employee attrition is its cost to an organization. Job postings, hiring processes, paperwork, and new hire training are some of the common expenses of losing employees and replacing them. Additionally, regular employee turnover prohibits your organization from increasing its collective knowledge base and experience over time. This is especially concerning if your business is customer-facing, as customers often prefer to interact with familiar people. Errors and issues are more likely if you constantly have new workers.

```python
# Lets split the target variable and feature variable before removing the skewness
X = hr_df.drop(columns=["Attrition"])
y = hr_df[["Attrition"]]
```

```python
# Lets keep the threshold as 0.6 and remove the skewness using transformation
for index in X.skew().index:
    if X.skew().loc[index]>0.6:
        X[index]=np.log1p(X[index])
    if X.skew().loc[index]<-0.6:
        X[index]=np.square(X[index])
```

```python
# Rechecking for skewness
print(X.skew())
print(f"Numeric features: {len(X.skew())}")
print(f'Count of features with skewness after transforming :{len(X.skew().loc[abs(X.skew())>0.6])}')
```

```python
# Re-plot distribution plot after removing skewness
X.plot(kind='density', subplots = True, layout = (6,11), sharex = False, legend= False, fontsize =1, figsize = (18,12))
plt.show()
```

#### Algorithm training and testing

```python
# Importing the model libraries
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lg_hr = LogisticRegression()
gnb_hr = GaussianNB()
svc_hr = SVC()
dtc_hr = DecisionTreeClassifier()
knn_hr = KNeighborsClassifier()
rfc_hr = RandomForestClassifier()
gbc_hr = GradientBoostingClassifier()

model = [lg_hr, gnb_hr, svc_hr, dtc_hr, knn_hr, rfc_hr, gbc_hr]
```

```python
# Maximum accuracy
maxAccu_hr = 0 

# Best random state value for which accuracy is achieved 
maxRS_hr = 0
for m in model:
    for i in range(0,500):
        x_train, x_test, y_train, y_test= train_test_split(train_x,train_y,test_size= 0.2, random_state=i)
        m.fit(x_train, y_train)
        train_score_hr = m.score(x_train, y_train)
        test_score_hr = m.score(x_train, y_train)
        pred_hr = m.predict(x_test)
        print(f'----------------At random State {i} ------------------')
        acc_hr = accuracy_score(y_test, pred_hr)
        print("Accuracy score of ", m, 'is:', acc_hr)
        print(f'Training score of {m} is {train_score_hr}')
        print(f'Testing score of {m} is {test_score_hr}')
        print('\n')
    
        if acc_hr > maxAccu_hr:
            maxAccu_hr = acc_hr
            maxRS_hr = i
            print(f"Maximum accuracy score of {m} is {acc_hr} at Random state {i}")
            print('\n')
    print('*'*120)
```

##### Lets see what is Maximum accuracy for each Algorithm

1. Maximum accuracy score of LogisticRegression() is 0.8582995951417004 at Random state 118
2. Maximum accuracy score of GaussianNB() is less than LogisticRegression()
3. Maximum accuracy score of SVC() is 0.9352226720647774 at Random state 118
4. Maximum accuracy score of DecisionTreeClassifier() is less than SVC()
5. Maximum accuracy score of KNeighborsClassifier() is less than SVC()
6. Maximum accuracy score of RandomForestClassifier() is 0.9696356275303644 at Random state 248
7. Maximum accuracy score of GradientBoostingClassifier() is slightly less than RandomForestClassifier() i.e. 0.966024340770791

- Finally Random forest classifier is giving best accuracy of all the algorithms. Next best accuracy is given by Gradient Boosting Classifier.

#### Cross validation of all the Algorithms

```python
# Before proceeding with the hpyer parameter tunning we will check for cross validation
from sklearn.model_selection import cross_val_score
max_cvscore_hr = 0 
max_cv_hr = 0
for m in model:
    print(f'---------------------- Cross Validation of {m} --------------------------')
    for j in range(2,15):
        cv_score_hr = cross_val_score(m, X,y,cv=j)
        cv_mean_hr = cv_score_hr.mean()
        print(f"At cross fold {j} the cv score is {cv_mean_hr}")
        print('\n') 
    
        if cv_mean_hr > max_cvscore_hr:
            max_cvscore_hr=cv_mean_hr
            max_cv_hr = j
            print(f"At cross fold {j} the Maximum CV score is {max_cvscore_hr}")
            print('\n') 
    print('*'*120)
    print('\n')
```

#### Best Cross validation for each algorithm
1. LogisticRegression() - At cross fold 10 the Maximum CV score is 0.8755102040816325
2. GaussianNB() - Less than LogisticRegression()
3. SVC() - Less than LogisticRegression()
4. DecisionTreeClassifier() - Less than LogisticRegression()
5. KNeighborsClassifier() - Less than LogisticRegression()
6. RandomForestClassifier() - Less than LogisticRegression()
7. GradientBoostingClassifier() - Less than LogisticRegression()

- We can see that Logistic regression is giving best cross validation score at cv 10. Since random forest is giving good accuracy score we will proceed with Random forest (Best CV at 14) for final deployment.
- Before deploying we will carry out hyperparameter tunning and check classification report, confusion matrix and AUC ROC curve.

#### Training each algorithm for Random state 248

```python
for m in model:
    x_train, x_test, y_train, y_test= train_test_split(train_x,train_y,test_size= 0.2, random_state=248)
    m.fit(x_train, y_train)
    train_score_hr = m.score(x_train, y_train)
    test_score_hr = m.score(x_train, y_train)
    pred_hr = m.predict(x_test)
    print(f'--------------------- {m} ----------------------')
    acc_hr = accuracy_score(y_test, pred_hr)
    print("Accuracy score of ", m, 'is:', acc_hr)
    print(f'Training score of {m} is {train_score_hr}')
    print(f'Testing score of {m} is {test_score_hr}')
    print("Confusion Matrix:\n",confusion_matrix(y_test, pred_hr))
    print('Classification Report:\n',classification_report(y_test, pred_hr))
    print('*'*120,'\n')
```

##### We can clearly see that Random Forest calssifier is giving the best results and the next algorithm with best results is Gradien Boosting Classifier

### Hyper Parameter Tunning 

```python
from sklearn.model_selection import GridSearchCV
```

##### Tunning with Logistic Regression

```python
param_lg = { 'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
             'multi_class': ['auto', 'ovr', 'multinomial'],
             'penalty': ['l1', 'l2', 'elasticnet'],
              'C':[100, 10, 1.0, 0.1, 0.01]}

gcv_lg_hr = GridSearchCV(lg_hr,param_lg,cv=10,scoring= 'accuracy')

# training
gcv_lg_hr.fit(x_train, y_train)

# Getting the best parameters
print("Best parameters:", gcv_lg_hr.best_params_)

print("Best Estimator:",gcv_lg_hr.best_estimator_)

gcv_lg_pred_hr=gcv_lg_hr.best_estimator_.predict(x_test)

print('Final Accuracy with Logistic Regression:', accuracy_score(y_test,gcv_lg_pred_hr))
```
